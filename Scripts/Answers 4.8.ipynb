```python
#Import Libraries
import pandas as pd
import numpy as np
import os
```

```python
ords_prods_merge= pd.read_pickle(r'/Users/mattrauth/Desktop/Instacart Basket Analysis /Data/Prepared Data/orders_products_merged_with_new_variable.pkl')
```

```python
#Create subset of first million rows
df = ords_prods_merge[:1000000]
```

```python
#Check shape of subset
df.shape
```

```python
df.head(10)
```

```python
#Split data into groups (with group being “product_name” column)
df.groupby('product_name')
```

```python
#Split the data into groups based on “department_id.” & Aggregate each group for mean values of “order_number” column
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
#Calculating mean w/out agg() function
df.groupby('department_id')['order_number'].mean()
```

```python
#Perform Multiple Aggregations

df.groupby('department_id').agg({'order_number': ['mean', 'min', 'max']})
```

```python
#Aggregating with transform()

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
#Tell pandas not to assign options regarding max # of rows to display
pd.options.display.max_rows = None
```

```python
ords_prods_merge.head(100)
```

```python
#Deriving Columns with loc()

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
#Check frequency

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```
